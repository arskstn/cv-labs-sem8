# 3.1 Получить видеопоток с веб-камеры

```python
cap = cv2.VideoCapture(0)

frames_sequence = []

for i in range(10):
    ret, frame = cap.read()
    if ret:
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frames_sequence.append(frame_rgb)
    
    cv2.waitKey(100) 

cap.release()

frame_initial = frames_sequence[0]
frame_test = frames_sequence[7]

f, axarr = plt.subplots(1, 2, figsize=(12, 6))
axarr[0].imshow(frame_initial)
axarr[0].set_title('Начальный кадр (Фон)')

axarr[1].imshow(frame_test)
axarr[1].set_title('Кадр с движением')
plt.show()
```
